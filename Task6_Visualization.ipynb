```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Downloads\2015.csv")
```

```python
df.head()
df.info()
```

The bar chart shows that countries with the highest happiness scores are concentrated at the top, indicating a clear gap between top-performing and lower-ranked countries.

```python
#Bar Chart â€“ Top 10 Happiest Countries
top10 = df.sort_values(by='Happiness Score', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10['Country'], top10['Happiness Score'])
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.title('Top 10 Happiest Countries')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

The line chart reveals a steady decline in happiness score as the happiness rank increases, showing a strong inverse relationship between rank and score.

```python
#Line Chart â€“ Happiness Rank vs Score
sorted_df = df.sort_values(by='Happiness Rank')

plt.figure(figsize=(8,5))
plt.plot(sorted_df['Happiness Rank'], sorted_df['Happiness Score'])
plt.xlabel('Happiness Rank')
plt.ylabel('Happiness Score')
plt.title('Happiness Score by Rank')
plt.grid(True)
plt.show()
```

```python
#Histogram â€“ Distribution of Happiness Scores
plt.figure(figsize=(8,5))
plt.hist(df['Happiness Score'], bins=15)
plt.xlabel('Happiness Score')
plt.ylabel('Number of Countries')
plt.title('Distribution of Happiness Scores')
plt.grid(True)
plt.show()
```

The scatter plot indicates a positive correlation between GDP per capita and happiness score, suggesting that higher economic prosperity generally contributes to higher happiness.

```python
#Scatter Plot â€“ GDP vs Happiness Score
plt.figure(figsize=(8,5))
plt.scatter(df['Economy (GDP per Capita)'], df['Happiness Score'])
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('GDP per Capita vs Happiness Score')
plt.grid(True)
plt.show()
```
